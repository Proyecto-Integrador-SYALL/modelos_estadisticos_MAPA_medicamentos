```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN



%matplotlib inline
sns.set()
```

```python
data_frame = pd.read_csv('data/standscaler_pca_dataset.csv', index_col='patient_id')
data_frame.head()
```

# DBSCAN

```python
clustering = DBSCAN(eps=3, min_samples=2).fit_predict(data_frame)
data_frame['cluster_dbscan'] = clustering

data_frame.head()
```

```python
fig, axes = plt.subplots(
    nrows=1, 
    ncols=3,
    figsize=(18,6)
)

_ = sns.scatterplot(
    x="0",
    y="1",
    data=data_frame,
    hue='cluster_dbscan',
    palette='Paired', 
    ax=axes[0]
)

_ = sns.scatterplot(
    x="0",
    y="2",
    data=data_frame,
    hue='cluster_dbscan',
    palette='Paired', 
    ax=axes[1]
)

_ = sns.scatterplot(
    x="2",
    y="1",
    data=data_frame,
    hue='cluster_dbscan',
    palette='Paired', 
    ax=axes[2]
)
```
