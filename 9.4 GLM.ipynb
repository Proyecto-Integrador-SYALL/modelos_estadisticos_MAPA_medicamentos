```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sn
%matplotlib inline
```

# Lectura de los datos.

```python
dfgrande=pd.read_csv('/Users/lescobar/Desktop/Proyecto Integrador/data/filtered1_dataset.csv', index_col="patient_id")
dfgrande.shape
```

# Ajuste de las columnas y variables dummy.

```python
dfgrande = dfgrande[['height','weight','age','sistolic','diastolic','heart_rate','pulse_pressure','cardiac_index','systemic_vascular_resistance','treatment','gender','smoking','excercise_frecuency','beer','wine','alcohol','strong_licor']]
dfgrande.head()
```

```python
#variables dummy.. 

genero=pd.get_dummies(dfgrande['gender'])
merged= pd.concat([dfgrande, genero], axis=1)

dfgrande=merged.drop(['gender', 'm'], axis=1)
dfgrande.head()
```

# Resumen estadístico.

```python
# resumen estadistico con pandas
dfgrande.describe()# correlacion
```

# Normalizar las variables numéricas

```python
cols = list(dfgrande.columns)
cols.remove('f')
cols.remove('treatment')
cols.remove('excercise_frecuency')
cols.remove('beer')
cols.remove('wine')
cols.remove('alcohol')
cols.remove('smoking')
cols.remove('strong_licor')

df_g=dfgrande[cols]
```

```python
from sklearn import preprocessing
# obtener los nombres de las columnas
names = df_g.columns
# Crear el objeto escalador
scaler = preprocessing.StandardScaler()
# Ajustar datos en el objeto escalador
scaled_df = scaler.fit_transform(df_g)
scaled_df = pd.DataFrame(scaled_df, index=dfgrande.index, columns=names)
scaled_df.head()
```

```python
scaled_df.shape
```

```python
# se unen todas las variables
df_num = scaled_df[['height','weight','age','sistolic','diastolic','heart_rate','pulse_pressure','cardiac_index','systemic_vascular_resistance']]
df_cat = dfgrande[['smoking','excercise_frecuency','beer','wine','alcohol','strong_licor', 'f' ]]
print(df_num.shape)
print(df_cat.shape)
```

```python
df_Y = dfgrande[['treatment']]
```

```python
df_X = pd.concat([df_num, df_cat], axis=1)
df_X.head()
```

```python
# Dividir el dataset en train y test
X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, test_size=0.25, random_state=0)
```

```python
import statsmodels.api as sm
model_bin = sm.GLM(y_train, X_train, family=sm.families.Binomial())
bin_results = model_bin.fit()
print(bin_results.summary())
```
