# Regresión Lineal

 

 

## Introducción

 

Los modelo de regresión lineal múltiple estudian la relación entre una variable de interés $Y$  la cual conoceremos como la variable respuesta o dependiente y un conjunto de variables explicativas o regresoras denotadas $X1, X2, . . . , Xp$.

 

En el modelo de regresión lineal múltiple se supone que la función de regresión que relaciona la variable dependiente con las variables independientes es lineal, es decir:

 

$$ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + · · · + \beta_pX_p + \epsilon $$

 

Donde: 
- $\beta_0$ es el término independiente. Es el valor esperado de $Y$ cuando $X_1, . . . , X_p$ son cero.
- $\beta_1, \beta_2, . . ., \beta_p$ son los coeficientes parciales de la regresión:
    - $\beta_1$ mide el cambio en $Y$ por cada cambio unitario en $X_1$, manteniendo $X_2, X_3, . . . , X_p$ constantes.
    - $\beta_2$ mide el cambio en $Y$ por cada cambio unitario en $X_2$, manteniendo $X_2, X_3, . . . , X_p$ constantes.
    - ...
    - $\beta_p$ mide el cambio en $Y$ por cada cambio unitario en $X_p$ , manteniendo $X_2, X_3, . . . , X_{p-1}$ constantes.
- $ε$ es el error de observación debido a variables no controladas.

 

El objetivo será obtener a partir de la muestra estimadores:
- De los coeficientes $\hat{\beta_0}, \hat{\beta_1}, . . . , \hat{\beta_p}$
- De la varianza del error $\hat{σ^2}.$

 

La información muestral se verá de la siguiente manera:

 

$$\begin{pmatrix}
y_{11} & x_{12} & ... & x_{1p}  \\
y_{21} & x_{22} & ... & x_{2p}  \\
... & & &\\
y_{n1} & x_{n2} & ... & x_{np}
\end{pmatrix}$$

 

Matricialmente el modelo de regresión lineal multiple se puede expresar de la siguiente manera:

 

$$\begin{pmatrix}
y_{11}\\
y_{21}\\
...\\
y_{n1}
\end{pmatrix} = \begin{pmatrix}
x_{12} & ... & x_{1p}  \\
x_{22} & ... & x_{2p}  \\
... & &\\
x_{n2} & ... & x_{np}
\end{pmatrix}  .  \begin{pmatrix} 
\beta_0\\
\beta_1\\
...\\
\beta_p 
\end{pmatrix} + \begin{pmatrix}
ε1\\
ε2\\
...\\
εn
\end{pmatrix} $$

 

Asignando la notación a las matrices anteriores, la expresión anterior se podrá escribir:
$$Y = X\beta + \epsilon$$

 

- Utilizando mínimos cuadrados para estimar el vector de parámetros $β$, se tiene que el estimador es:

 

$$ \hat{\beta} = (X^t X)^{−1}X^tY $$
donde $X^t$ denota a la matriz transpuesta de $X$.

 

- El estimador de la varianza del error se puede definir como:

 

$$ σ^2 = \frac{1}{n − (p + 1)} \sum_{i=1}^{n} \epsilon^2 = \frac{1}{n − (p + 1)} \sum_{i=1}^{n}(y_i − \hat{y_i})^2$$

 

Descomposición de la variabilidad

 

- La varibilidad toda la muestra se denomina variabilidad total (VT).

 

$$ VT=\sum_{i=1}^{n}(y_i − \bar{y_i})^2$$

 

Tambien es posible descomponerla en dos sumandos:

 

- La variabilidad explicada (VE), dado por la regresión con p grados de libertad.

 

$$ VE=\sum_{i=1}^{n}(\hat{y_i} − \bar{y_i})^2$$

 

- La variabilidad no explicada (VNE , dado por los residuales con n-(p+1) grados de libertad

 

$$ VNE=\sum_{i=1}^{n}(y_i − \hat{y_i})^2$$

 

Finalmente,

 

$$VT= VE+VNE$$ con n-1 grados de libertad 

 

## Significancia e intervalo de confianza para los $β's$

 

Luego de calcular el modelo es importante validar si realmente esta siendo explicativo, es decir, las variables explicativas si tienen algún efecto sobre la variable respuesta. Esto se realiza mediante pruebas de hipotesis, luego de verificar que se cumplan los supuestos del modelo.

 

$$
\left\{
\begin{array}{ll}
H_{0}: &  \beta_1=\beta_2=...=\beta_p=0\\
H_{1}: & \beta_j \neq 0   \text{para algun j=1,...,p}
\end{array}
\right.
$$

 

- Si no rechazamos la hipótesis nula el modelo no es explicativo, es decir, ninguna variables explicativa influye en la variable respuesta $Y$.

 

- Si rechazmos la hipótesis nula, el modelo es explicativo, es decir, al menos una de las variables explicativa influye en la variable respuesta $Y$.

 

Y finalmente se necesitará determinar cual es el conjunto de variables explicativas: $x_1, x_2, …, x_k$, que influyen en la variable dependiente $Y$, al igual que anterirmente se realiza la prueba de hipótesis para cada coeficiente $_\beta$

 

$$
\left\{
\begin{array}{ll}
H_{0}: &  \beta_j=0  \text{x_j no influye sobre Y}\\
H_{1}: & \beta_j \neq 0   \text{x_j influye sobre Y}
\end{array}
\right.
$$

 


## Residuos del modelo

 

El residuo de una estimación se define como la diferencia entre el valor observado y el valor esperado acorde al modelo. 

 

Cuanto mayor es el sumatorio del cuadrado de los residuos menor la precisión con la que el modelo puede predecir el valor de la variable dependiente a partir de la variable predictora. Los residuos son muy importantes puesto que en ellos se basan las diferentes medidas de la bondad de ajuste del modelo.

 

## Bondad de ajuste del modelo

 

Una vez que se ha ajustado un modelo es necesario verificar su eficiencia, ya que aun siendo la línea que mejor se ajusta a las observaciones de entre todas las posibles, el modelo puede ser malo. Para este caso haremos uso del 
*coeficiente de determinación* $R^2$.

 

**Coeficiente de determinación** $R^2$: Describe la proporción de variabilidad observada en la variable dependiente $Y$ explicada por el modelo y relativa a la variabilidad total. Su valor está acotado entre 0 y 1.

 

$$R^2= \frac{VE}{VT} = 1-\frac{VNE}{VT}$$

 

El coeficiente de determinación presenta el inconveniente de aumentar siempre que aumenta el número de variables regresoras, por ello se introduce una penalización del número de variables regresoras que se incluyen en el modelo de regresión, es conveniente utilizar el coeficiente de determinación corregido por el número de grados de libertad.

 

$$ R^2_{ajustado}=1- \frac{VNE/(n-(p+1))}{VT/(n-1)}$$

 

## Supuestos para la regresión lineal

 

- **Linealidad**: La relación entre las variables explicativas debe ser lineal. 

 

- **Distribución Normal de los residuos**: Los residuos se tiene que distribuir de forma normal, con media igual a 0. Esto se puede comprobar con un histograma, con la distribución de cuantiles (qqnorm() + qqline()) o con un test de hipótesis de normalidad. Los valores extremos suelen ser una causa frecuente por la que se viola la condición de normalidad.

 


- **Varianza de residuos constante** (homocedasticidad): La varianza de los residuos ha de ser aproximadamente constante a lo largo del eje X. Se puede comprobar mediante gráficos (scatterplot) de los residuos de cada observación (formas cónicas son un claro indicio de falta de homocedasticidad) o mediante contraste de hipótesis mediante el test de Breusch-Pagan.

 


- **Valores atípicos y de alta influencia**: Hay que estudiar con detenimiento los valores atípicos o extremos ya que pueden generar una falsa correlación que realmente no existe, u ocultar una existente. (Ver descripción detallada en la sección de apuntes varios).

 


- **Independencia, Autocorrelación**: Las observaciones deben ser independientes unas de otras. Esto es importante tenerlo en cuenta cuando se trata de mediciones temporales. Puede detectarse estudiando si los residuos siguen un patrón o tendencia. Otro caso frecuente es el de tener varias mediciones para un mismo sujeto. En estos casos, primero se obtiene la media de cada uno y después se ajusta el modelo empleando las medias.

 

Dado que las condiciones se verifican a partir de los residuos, primero se suele generar el modelo y después se valida. De hecho, el ajuste de un modelo debe verse como un proceso iterativo en el que se ajusta un modelo inicial, se evalúa mediante sus residuos y se mejora. Así hasta llegar a un modelo óptimo.
